```python
df= df.drop("Music",axis = 1)
df
```

```python
df['Gender']=df['Gender'].replace({0:"Male",1:"Female"})
df
```

# Gender Distribution

```python
a = sns.countplot(data = df, x= "Gender")

plt.title("Gender Distribution")
plt.show()
a.bar_label(a.containers[0])
```

```python
# FROM THE ABOVE CHART WE HAVE ANALYSED THAT
# NO. OF FEMALES IS MORE THAN THE NO. OF MALES.
```

# Realtionship Between Gender with Weeklystudy time , Absences and Gradeclass

```python
B = df.groupby("Gender").agg({"StudyTimeWeekly":'mean',"Absences":'mean',"GradeClass":'mean'})
B
```

```python
sns.heatmap(B,annot = True)

plt.title(" Relation between Genders and Studytime , Absences and GradeClass")
plt.show()
```

```python
# FROM THE ABOVE WE HAVE ANALYSED THAT
# FEMALES ARE MORE ABSENT AS COMPARED TO MALES 
# THEIR IS NO/NEGLIGABLE DIFFERENCE IN THEIR WEEKLY STUDY TIME.
```

# Ethinicity

```python
df["Ethnicity"].unique()
```

```python
df['Ethnicity']=df['Ethnicity'].replace({0:"null",1:"group1",2:"group2",3:"group3"})
df
```

```python
One=df.loc[(df["Ethnicity"] == 'group1')].count()
Two=df.loc[(df["Ethnicity"] == 'group2')].count()
Three=df.loc[(df["Ethnicity"] == 'group3')].count()

m=[One["Ethnicity"],Two["Ethnicity"],Three["Ethnicity"]]
l = ("Group 1","Group 2","Group 3")

plt.title("DIstribution of Ethnicity")
plt.pie(x=m,labels = l, autopct= "%1.2f%%")
plt.show()
```
